Now, the same for [itch.io](itch.io). Again, only open the cells if you need help. How would you bring the data into Python?

```python
# @title
import requests

url = "https://itch.io/games?page=33&format=json"
response = requests.get(url)

response.status_code
```

```python
# @title
games_data = response.json()
type(games_data)
```

When I found the data, it was in a dictionary. Given a dictionary of game information, let's scrub through it. With dictionaries you look for their keys...

```python
games_data.keys()
```

```python
games_data['num_items']
```

It turns out that one of the keys has HTML content for the particular chunk of games (chunks appear as you scroll down the page). There's a hook to even iterate through the chunks programmatically.

```python
from bs4 import BeautifulSoup

games_page = BeautifulSoup(games_data["content"])
print(games_page.prettify())
```

Compare finding information in this HTML code versus in a dictionary. How are things different? What do you have to look to for to find the information you're after?

Here we appeal to tag name and tag class. BeautifulSoup lets us search HTML for tags with certain attributes. It also can make use of a CSS selector directly. Here we just look for all the `<a>` anchor tags of class "title".

```python
games = games_page.find_all("a","title")
len(games)
```

```python
games[10]
```

```python
type(games[0])
```

```python
game = games[0]

print(game.prettify())
```

```python
game.get_text()
```

```python
game["href"]
```

And here we look for all the `<div>` tags with class "game_cell_data".

```python
games = games_page.find_all("div","game_cell_data")
len(games)
```

```python
games[5]
```

```python
game = games[5]
print(game.prettify())
```

BeautifulSoup really is beautiful. You can take one of the games we've found and dig into it. It's just another tag, after all.

```python
game.find("div","game_genre")
```

```python
game.find("div","game_genre").get_text()
```

And here we iterate through the games, makings lists of game titles and game authors, eventually creating a DataFrame. We could imagine iterating this and getting all the games.

```python
who = []
what = []

for game in games:
  what.append(game.find("a","title").get_text())
  who.append(game.find("div","game_author").get_text())

import pandas as pd

game_df = pd.DataFrame({"author":who,"title":what})
game_df
```
